### Demonstrates quirky self-referencial behavior of tuple unpacking in chained assignments

```python
a, b = a[b] = {}, 5
assert a[b][0] == a[b][0][b][0][b][0]
a
```

<img src="https://i.imgur.com/DWrI2JY.gif" align="right" />

Explanation:
1. Python evaluates expressions from left to right.
2. But in an assignment, the right-side is evaluated first.
3. Therefore, `a` is set to `{}`, and `b` is set to `5`.
4. Then, `a[b]` is evaluated with the new values of `a` and `b`.
